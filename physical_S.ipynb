# 3.sonar(초음파센서_pin13)
### 핀모드 설정: rs.sonar_begin(pin)




```python
# 실행하세요
rs.sonar_begin(13)
time.sleep(1)
print('초음파센서 작동시작')
```

```python
#실행해 보세요. 거리가 가까우면 작은값, 멀어지면 값이 커집니다.
try:
    for i in range(100):
        sonar = rs.sonar_read(13)
        print(sonar,end=',')
        time.sleep(0.1)
    print('10초 테스트 끝')
except KeyboardInterrupt:
    print('중지 버튼을 눌렀습니다.')
```

<코딩하기>
- 초음파센서를 이용하여 거리에 따라 도레미파솔~ 음이 나오도록 코딩해 보세요
- 센서값 읽기 함수 : rs.sonar_read(pin)



```python
#여기에 코딩해 보세요
try :    
    
    while True :
        sonar = rs.sonar_read(13) 

        if 2 <= sonar <= 5 :
            buzzer(3,36,250)  #도
            time.sleep(0.5)
        elif 6 <= sonar <= 10 :  
            buzzer(3,38,250)  #레  
            time.sleep(0.5)
        elif 11<= sonar <= 15 :  
            buzzer(3,40,250)  #미  
            time.sleep(0.5)
        elif 16<= sonar <= 20 :
            buzzer(3,41,250)  #파    
            time.sleep(0.5)
        elif 21<= sonar <= 25 :
            buzzer(3,43,250)  #솔   
            time.sleep(0.5)
        elif 26<= sonar <= 30 :
            buzzer(3,45,250)  #라    
            time.sleep(0.5)
        
except KeyboardInterrupt:
    print('중지 버튼을 눌렀습니다')
```

# 5. 최종미션
- 시작 : 초음파센서 작동중 출력(print)
- 차단기 열기(서보모터)
- 초음파센서 계속 작동 
- 만약 도둑이 침입했다면(sonar <10)
- 차단기 내려오기
- 도둑 침입 경찰출동 출력(print,7회반복)
- 경보음 / led 켜지기(7회반복)
- break 



```python
#여기에 코딩해 보세요.
try:
    servo(12,30)
    print('초음파센서 작동중')
    while True :
        sonar  = rs.sonar_read(13)
        
        if sonar < 10 :          
            servo(12,-60)
            time.sleep(0.1)
            for i in range(7):
                print('도둑이 침입 했습니다!경찰출동')
                led_color(2,100,0,0)
                led_on(2)
                buzzer(3,66,250)                
                led_off(2)
                time.sleep(0.5)
            break  
                
except KeyboardInterrupt:
   
    print('중지')
                                                            
```
